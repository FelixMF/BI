```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

data = pd.read_csv("factVentasEtiquetados.csv", sep=';' ,encoding='ISO-8859-1')
```

```python

columnas_numericas = data.select_dtypes(include=[np.number]).columns
for columna in columnas_numericas:
    datos = data[columna]
    mu = np.mean(datos)
    sigma = np.std(datos)

    x = np.linspace(min(datos), max(datos), 1000)

    pdf = norm.pdf(x, mu, sigma)

    cdf = norm.cdf(x, mu, sigma)

    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(x, pdf, label='PDF')
    plt.title('Función de Densidad de Probabilidad (PDF) - ' + columna)
    plt.xlabel(columna)
    plt.ylabel('Densidad de Probabilidad')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(x, cdf, label='CDF', color='orange')
    plt.title('Función de Distribución Acumulativa (CDF) - ' + columna)
    plt.xlabel(columna)
    plt.ylabel('Probabilidad Acumulada')
    plt.legend()

    plt.show()
```
